```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("customerchurn.csv")

print("Total Customers:", len(df))
print("\nColumns in Dataset:")
print(df.columns.tolist())

churn_column = input("\nEnter the churn column name exactly as shown above: ")

print("\nCustomer Status Count:")
print(df[churn_column].value_counts())

churn_customers = df[df[churn_column].astype(str).str.lower().isin(["yes", "churned", "1"])].shape[0]

total_customers = len(df)
churn_rate = (churn_customers / total_customers) * 100
retention_rate = 100 - churn_rate

print(f"\nChurn Rate: {churn_rate:.2f}%")
print(f"Retention Rate: {retention_rate:.2f}%")

df[churn_column].value_counts().plot(kind="bar")
plt.title("Customer Churn Analysis")
plt.xlabel("Status")
plt.ylabel("Number of Customers")
plt.show()

if "MonthlyCharges" in df.columns:
    print("\nAverage Monthly Charges:")
    print(df.groupby(churn_column)["MonthlyCharges"].mean())

if "tenure" in df.columns:
    print("\nAverage Tenure:")
    print(df.groupby(churn_column)["tenure"].mean())

print("\n--- Insights ---")
print("1. Customer distribution analyzed successfully.")
print("2. Churn and retention rates calculated.")
print("3. Retention strategies can target customers likely to churn.")
```
